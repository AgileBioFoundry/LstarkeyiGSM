load other GSMs.

```python

# Rhodosporidium toruloides IFOxxxx.
rto = cobra.io.load_json_model('../models/Rt_IFO0880.json')

# Yarrowia lipolytica POxx.
# ylip = cobra.io.load_matlab_model('../models/twoModels/iYLI647_corr.mat')

# Sacchromyces cerevisiae Yeast8.
sce = cobra.io.load_matlab_model('../models/yeast8_modifiedwBIGGnames.mat')


# E. coli iJO1366
eco = cobra.io.load_json_model('../models/iJO1366.json')
```

ensuring reasonable bounds on reactions.

```python
for r in model.reactions:
    if r.upper_bound>1000:
        
        print(r.id,r.bounds)
        r.upper_bound = 1000
    if r.lower_bound<-1000:
        print(r.id,r.bounds)
        r.lower_bound = -1000
```

reexaming subsystems.

```python
subsystems=[]
rx_id = []
for rx in model.reactions:
    sub=rx.subsystem
    subsystems.append(sub)
    rx_id.append(rx.id)
rx_sub_dict = dict(zip(rx_id,subsystems))
subs = set(subsystems)
```

```python
subs
```

Finding reactions with no assigned subsystem. 

```python
keys_for_value = [key for key, value in rx_sub_dict.items() if value == '']
keys_for_value
```

```python
for rxn in keys_for_value:
    rx = model.reactions.get_by_id(rxn)
    print(rx.id,rx.name,rx)
```

```python
len(keys_for_value)
```

load other GSM to determine subsystems.


```python
rxn_new_subsystems = dict()
for reaction_id_to_check in keys_for_value:
    if reaction_id_to_check in rto.reactions:
        sub = rto.reactions.get_by_id(reaction_id_to_check).subsystem
        if sub:
            rxn_new_subsystems[reaction_id_to_check]=sub
    if reaction_id_to_check in sce.reactions:
        print(reaction_id_to_check,sce.reactions.get_by_id(reaction_id_to_check).subsystem)
        sub = sce.reactions.get_by_id(reaction_id_to_check).subsystem
        if sub:
            rxn_new_subsystems[reaction_id_to_check]=sub   
    if reaction_id_to_check in eco.reactions:
        print(reaction_id_to_check,eco.reactions.get_by_id(reaction_id_to_check).subsystem)
        sub = eco.reactions.get_by_id(reaction_id_to_check).subsystem
        if sub:
            rxn_new_subsystems[reaction_id_to_check]=sub
```

```python
rxn_new_subsystems
```

```python
len(rxn_new_subsystems)
```

assigning 16 new subsystems.

```python
for rxid in list(rxn_new_subsystems.keys()):
    model.reactions.get_by_id(rxid).subsystem = rxn_new_subsystems[rxid]
    print(rxid,model.reactions.get_by_id(rxid).subsystem)
```
