# GSM_change_yeast8.

This notebook loads the yeast8 GSM and changes the id conventions to the BiGG reaction annotated by the model (if there is one present). 


load libraries. 

```python
import cobra
import pandas as pd
import numpy as np
```

```python
# load the yeast8 GSM (.mat) file. 
yeast8 = cobra.io.load_matlab_model("../models/yeast-GEM.mat")
```

```python
yeast8
```

```python
yeast8.reactions
```

```python
# load the yeast8 excel data. this data contains the corresponding bigg reaction IDs in a column titled 'MIRIAM'.
yeast8Data = pd.read_excel("../models/yeast-GEM.xlsx")
```

```python
yeast8Data
```

```python
# obtain the BiGG annotations. 
temp = yeast8Data.MIRIAM.to_list()
```

```python
# parse the annotations to get the 'ID'.
temp2 = [i.split(';')[0] for i in temp]
temp3 = [i.split('/')[1] for i in temp2]
temp4 = [i.split('/')[0] for i in temp2]
```

```python
# dictionary storing the reaction name and the reaction source. 
dict_rxn_source = dict(zip(temp3,temp4))
```

```python
dict_rxn_source
```

```python
# create a copy of the yeast8 GSM that will be modified. 
yeast8_bigg = yeast8.copy()
```

```python
# obtain the rxn IDs in the GSM. 
rxns = [r.id for r in yeast8_bigg.reactions]
```

```python
dict_rxn_descriptName = dict(zip(rxns,temp3))
```

```python
# loop through reactions names in GSM and change the IDs to the bigg format. 
for r in yeast8_bigg.reactions:
    
    # if the source for the reaction name is from the BiGG database modify the name. 
    if dict_rxn_source[dict_rxn_descriptName[r.id]] == 'bigg.reaction':
        try:
            r.id = dict_rxn_descriptName[r.id]
        except:
            print('{} {} in model already'.format(r.id,dict_rxn_descriptName[r.id]))
            r.id = dict_rxn_descriptName[r.id] + '_'
```

```python
# load the yeast8 GSM from the Excel sheet. 
yeast8Data = pd.read_excel("../models/yeast-GEM_metsSheet.xlsx")
```

```python
yeast8Data.head()
```

```python
# obtain the BiGG naming convention. 
temp = yeast8Data.MIRIAM.to_list()
```

```python
# parse the columns to grab the BiGG met ID. 
temp2 = [i.split(';')[0] for i in temp]
temp3 = [i.split('/')[1] for i in temp2]
temp4 = [i.split('/')[0] for i in temp2]
```

```python
# obtain the metabolite compartments. 
temp_compartment = yeast8Data.COMPARTMENT.to_list()
```

```python
# modify the ids to add the compartment. 
temp5=[]
for i,e in enumerate(temp3):
    temp5.append(str(e)+'_'+str(temp_compartment[i]))
temp5
```

```python
# create a dictionary for mapping the new IDs. 
dict_met_source = dict(zip(temp5,temp4))
```

```python
dict_met_source
```

```python
# create a copy of the GSM for modification. 
# yeast8_bigg = yeast8.copy()
```

```python
# grab the metabolite IDs in the GSM. 
mets = [m.id for m in yeast8_bigg.metabolites]
```

```python
# dictionary for mapping. 
dict_met_descriptName = dict(zip(mets,temp5))
```

```python
# loop through model and modify the naming convention. 
for m in yeast8_bigg.metabolites:
    if dict_met_source[dict_met_descriptName[m.id]] == 'bigg.metabolite':
        try:
            m.id = dict_met_descriptName[m.id]
        except:
            print('{} {} in model already'.format(m.id,dict_met_descriptName[m.id]))
            m.id = dict_met_descriptName[m.id] + '_'
```
